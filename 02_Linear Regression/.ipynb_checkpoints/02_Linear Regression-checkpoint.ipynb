### 1.Generate simulated data

```python
import numpy as np
import matplotlib.pyplot as plt
X = 2 * np.random.rand(100,1) # Feature X
y = 4 + 3 * X + np.random.rand(100,1) # Real world: y = 4 + 3x + noise
y1 = 4 + 3 * X # No noise
plt.scatter(X, y, alpha=0.7, label="Original data")
```

```python
X_b = np.c_[np.ones((100,1)), X] # Add X0=1
```

### 2.Normal Equation

```python
theta_hat = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y) # np.linalg.inv() calculate inverse matrix, dot() calculate the dot product
theta_hat
```

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)), X_new]
y_predict = X_new_b.dot(theta_hat)
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y1, "g-")
plt.plot(X, y, "b.")
```

### 3.SVD

```python
U, S, Vt = np.linalg.svd(X_b, full_matrices=False)
theta_hat_1 = Vt.T @ np.diag(1 / S) @ U.T @ y
theta_hat_1
```

```python
y_predict_1 = X_new_b.dot(theta_hat_1)
plt.plot(X_new, y_predict, "r-")         # Normal Equation
plt.plot(X_new, y_predict_1, "g--")      # SVD
plt.plot(X, y, "b.")                     # Simulated data
```

### 4.scikit-learn

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X, y)
linreg.intercept_, linreg.coef_
y_predict_2 = linreg.predict(X_new)
plt.plot(X_new, y_predict_2, "b-")
plt.plot(X, y, "r.")
```
